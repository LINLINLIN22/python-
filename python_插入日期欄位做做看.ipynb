```python
files = glob('D:/python資料科學概論/year20/1/*.csv') # 欲合併的csv資料檔
print(files)
```

```python
# assign path
path, dirs, files = next(os.walk("D:/python資料科學概論/year20/1/"))
file_count = len(files)
# create empty list
dflist = []
  
# append datasets to the list 
for i in range(file_count):
        temp_df = pd.read_csv("D:/python資料科學概論/year20/1/"+files[i])
        dflist.append(temp_df) 
print(files,len(files))
```

```python
# rename columns & change to data frame & keep varible
for i in range(10):
    dflist[i].rename(columns = {'Province/State':'Province_State',
                                'Country/Region' : 'Country_Region','Last Update' : 'Last_Update'}, inplace = True)
    exec(f"df{i}=pd.DataFrame(dflist[i])") # change to data frame
    exec(f"df{i} = df{i}[['Province_State','Country_Region','Last_Update', 'Confirmed','Deaths','Recovered']]") # keep variable
```

```python
# let Year, month, day & put na to 0 & let Date & drop Year, month and day
for i in range(10):
    exec(f"df{i}.insert(3,column='Year',value='2020')") # let Year
    exec(f"df{i}.insert(4,column='month',value='01')") # let month 
    j=i+22 # let day
    exec(f"df{i}.insert(5,column='day',value='{j}')") 
    exec(f"df{i}['Confirmed'] = df{i}['Confirmed'].replace(np.nan, 0)") #na to 0
    exec(f"df{i}['Deaths'] = df{i}['Deaths'].replace(np.nan, 0)")
    exec(f"df{i}['Recovered'] = df{i}['Recovered'].replace(np.nan, 0)")
    exec(f"df{i}['Year'] = df{i}['Year'].astype(str)") # let Date
    exec(f"df{i}['month'] = df{i}['month'].astype(str)") 
    exec(f"df{i}['day'] = df{i}['day'].astype(str)")
    exec(f"df{i}['Date'] = df{i}[['Year', 'month','day']].apply('-'.join, axis=1)")
    exec(f"df{i}['Date'] = pd.to_datetime(df{i}['Date'], format='%Y-%m-%d')") # let Date to DateTime
    exec(f"df{i}=df{i}.drop(['Year','month','day'],axis=1)") # drop Year, month and day

# to 1 csv
for i in range(1,10):
    exec(f"df = df{i}=pd.concat([df{i}, df{i-1}], ignore_index=True).sort_values('Date')")
print(df)
df.to_csv("2020-01.csv", index=False, encoding="utf8")
```
